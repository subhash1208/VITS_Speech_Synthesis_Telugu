### Step 1: Preprocess Audio (Resampling & Conversion)

```python
import librosa
import soundfile as sf
import os

# Define the path to the WAV directory and the output directory for processed audio
wavs_dir = "D:\\Projects\\VITS_Speech_Synthesis\\data\\IndicVoices-R\\Telugu\\wavs"
output_dir = "D:\\Projects\\VITS_Speech_Synthesis\\data\\IndicVoices-R\\Telugu\\wavs_resampled"

def preprocess_audio(wavs_dir, output_dir, sample_rate=16000):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for wav_file in os.listdir(wavs_dir):
        if wav_file.endswith(".wav"):
            wav_path = os.path.join(wavs_dir, wav_file)
            output_path = os.path.join(output_dir, wav_file)

            # Load and resample the audio file
            audio, sr = librosa.load(wav_path, sr=sample_rate, mono=True)

            # Save the processed audio
            sf.write(output_path, audio, sample_rate)
            print(f"Processed: {wav_file}")

# Example usage
preprocess_audio(wavs_dir, output_dir)
```
